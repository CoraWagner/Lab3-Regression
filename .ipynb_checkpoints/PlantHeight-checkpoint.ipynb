```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score
import statistics as stat

file = pd.read_csv('Plant Height Data - Data.csv')

### Height Based on Temperature ###
height = file['height'].tolist()
temp = file['temp'].tolist()

# Plot data points
plt.scatter(temp, height)

# Plot line of best fit
x = np.array(temp)
y = np.array(height)
m, b = np.polyfit(x, y, 1)

plt.plot(x, y, 'o')
plt.plot(x, m*x + b)

# Add title and axis labels
plt.title("Plant Height Based on Temperature")
plt.xlabel("Temperature")
plt.ylabel("Height")

plt.show()

# R Squared
coeffs = np.polyfit(x, y, 1)
p = np.poly1d(coeffs)
y_hat = p(x)
y_bar = np.sum(y)/len(y)
ssreg = np.sum((y_hat-y_bar)**2)
sstot = np.sum((y - y_bar)**2)
rsquared = ssreg / sstot

print("R Squared for Plant Height Based on Temperature: " + str(rsquared))

### Height Based on Rain ###
rain = file['rain'].tolist()

# Plot data points
plt.scatter(rain, height)

# Plot line of best fit
x = np.array(rain)
m, b = np.polyfit(x, y, 1)  # Using the y from the original calculation (Didn't change durning calculations)

plt.plot(x, y, 'o')
plt.plot(x, m*x + b)

# Add title and axis labels
plt.title("Plant Height Based on Rain")
plt.xlabel("Rain")
plt.ylabel("Height")

plt.show()

# R Squared
coeffs = np.polyfit(x, y, 1)
p = np.poly1d(coeffs)
y_hat = p(x)
y_bar = np.sum(y)/len(y)
ssreg = np.sum((y_hat-y_bar)**2)
sstot = np.sum((y - y_bar)**2)
rsquared = ssreg / sstot

print("R Squared for Plant Height Based on Rain: " + str(rsquared))

### Height Based on Altitude ###
alt = file['alt'].tolist()

# Plot data points
plt.scatter(alt, height)

# Plot line of best fit
x = np.array(alt)
m, b = np.polyfit(x, y, 1)  # Using the y from the original calculation (Didn't change durning calculations)

plt.plot(x, y, 'o')
plt.plot(x, m*x + b)

# Add title and axis labels
plt.title("Plant Height Based on Altitude")
plt.xlabel("Altitude")
plt.ylabel("Height")

plt.show()

# R Squared
coeffs = np.polyfit(x, y, 1)
p = np.poly1d(coeffs)
y_hat = p(x)
y_bar = np.sum(y)/len(y)
ssreg = np.sum((y_hat-y_bar)**2)
sstot = np.sum((y - y_bar)**2)
rsquared = ssreg / sstot

print("R Squared for Plant Height Based on Altitude: " + str(rsquared))

### Height Based on Site Number ###
siteNum = file['site'].tolist()

# Plot data points
plt.scatter(siteNum, height)

# Plot line of best fit
x = np.array(siteNum)
m, b = np.polyfit(x, y, 1)  # Using the y from the original calculation (Didn't change durning calculations)

plt.plot(x, y, 'o')
plt.plot(x, m*x + b)

# Add title and axis labels
plt.title("Plant Height Based on Site Number")
plt.xlabel("Site Number")
plt.ylabel("Height")

plt.show()

# R Squared
coeffs = np.polyfit(x, y, 1)
p = np.poly1d(coeffs)
y_hat = p(x)
y_bar = np.sum(y)/len(y)
ssreg = np.sum((y_hat-y_bar)**2)
sstot = np.sum((y - y_bar)**2)
rsquared = ssreg / sstot

print("R Squared for Plant Height Based on Site Number: " + str(rsquared))
```
